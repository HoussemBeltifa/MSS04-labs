```python
def q_solve(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "all numbers"
            else:
                return "no real solution"
        else:
            return [-c / b]

    disc = b**2 - 4*a*c

    if disc < 0:
        return "no real solution"

    sqrt_d = disc ** 0.5

    x1 = (-b - sqrt_d) / (2*a)
    x2 = (-b + sqrt_d) / (2*a)

    if disc == 0:
        return [x1]

    return [x1, x2]
```

```python
print(q_solve(0, 0, 0))
print(q_solve(1, 6, 4))
print(q_solve(1, 2, 3))
```

```python
def p_learn(X, Y, w1, w2, b, eta):
    all_correct = False

    while not all_correct:
        all_correct = True

        for i in range(len(X)):
            x1 = X[i][0]
            x2 = X[i][1]
            y = Y[i]

            z = w1 * x1 + w2 * x2 + b
            y_pred = H(z)

            if y == 1 and z < 0:
                w1 = w1 + eta * x1
                w2 = w2 + eta * x2
                b = b + eta
                all_correct = False

            elif y == 0 and z >= 0:
                w1 = w1 - eta * x1
                w2 = w2 - eta * x2
                b = b - eta
                all_correct = False

    return w1, w2, b
```

```python
X = [[1, 1], [1, 0], [0, 1], [0, 0]]
Y_OR = [1, 1, 1, 0]

w1 = 0
w2 = 0
b = 0
eta = 1

w1, w2, b = p_learn(X, Y_OR, w1, w2, b, eta)

print("w1 =", w1)
print("w2 =", w2)
print("b =", b)

for x in X:
    print(x, "->", output(x[0], x[1], w1, w2, b))
```
