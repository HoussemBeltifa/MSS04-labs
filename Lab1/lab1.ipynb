```python
def q_solve(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "all numbers"
            else:
                return "no real solution"
        else:
            return [-c / b]

    disc = b**2 - 4*a*c

    if disc < 0:
        return "no real solution"

    sqrt_d = disc ** 0.5

    x1 = (-b - sqrt_d) / (2*a)
    x2 = (-b + sqrt_d) / (2*a)

    if disc == 0:
        return [x1]

    return [x1, x2]
```

```python
print(q_solve(0, 0, 0))
print(q_solve(1, 6, 4))
print(q_solve(1, 2, 3))
```

```python
import cmath


def q_solve_complex_number(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "all numbers"
            else:
                return "no solution"
        else:
            return [-c / b]

    disc = b**2 - 4*a*c

    sqrt_d = cmath.sqrt(disc)

    x1 = (-b - sqrt_d) / (2*a)
    x2 = (-b + sqrt_d) / (2*a)

    if x1 == x2:
        return [x1]

    return [x1, x2] 
```

```python
print(q_solve_complex_number(1, 0, 1))
print(q_solve_complex_number(1, -5, 6))  
print(q_solve_complex_number(1 + 1j, 2, 1))   
```
