```python
def q_solve(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "all numbers"
            else:
                return "no real solution"
        else:
            return [-c / b]

    disc = b**2 - 4*a*c

    if disc < 0:
        return "no real solution"

    sqrt_d = disc ** 0.5

    x1 = (-b - sqrt_d) / (2*a)
    x2 = (-b + sqrt_d) / (2*a)

    if disc == 0:
        return [x1]

    return [x1, x2]
```

```python
print(q_solve(0, 0, 0))
print(q_solve(1, 6, 4))
print(q_solve(1, 2, 3))
```

```python
def preceptron_learn(X, Y, w1, w2, b, eta):
    for epoch in range(3):
        print(f"\nepoch = {epoch+1}")
        for i in range(len(X)):
            a = output(X[i][0], X[i][1], w1, w2, b)
            y = Y[i]
            e = y-a
            print(X[i], f" y={y}, a={a}, e={e}")
            if y==1 and a ==0:
                w1 = w1 + eta * X[i][0]
                w2 = w2 + eta * X[i][1]
                b = b + eta
            elif y==0 and a==1:
                w1 = w1 - eta * X[i][0]
                w2 = w2 - eta * X[i][1]
                b = b - eta
            print(f"updated w1={w1}, w2={w2}, b={b}")
        updated_a = [output(X[i][0], X[i][1], w1, w2, b) for i in range(len(X))]
        print("_"*22)
        print(f"updated a={updated_a}, (y={y})")
```

C) OOP

```python
class perceptron:
    def __init__(self, n_inputs):
        self.n_inputs = n_inputs
        self.weights = np.zeros(n_inputs)
        self.bias = 0
        
    def out(self, X):
        X = np.array(X)
        z=np.dot(X, self.weights)+self.bias
        
        return 1 if z>=0 else 0
    
    def learn(self, data_learn, eta=0.1, epochs=10):
        err_list=[]
        for epoch in range(epochs):
            errors = 0
            for ix, y in data_learn: 
                a = self.out(ix)
                e = a-y
                if e != 0:
                    errors += 1
                if y==1 and a ==0:
                    self.weights = self.weights + eta * ix
                    self.bias = self.bias + eta
                elif y==0 and a==1:
                    self.weights = self.weights - eta * ix
                    self.bias = self.bias - eta
            err_list.append(errors)
            print(f"End of {int(epoch+1)}.epoch :")
            print(f"Errors per epoch = {errors}")
            print(f"Updated weights = {self.weights}, bias = {self.bias}")
            print("--"*20)
        return(err_list)
```

```python
data_learn = list(zip(X,Y_OR))
p = perceptron(2)
err_list = p.learn(data_learn)
print(f"List of errors per epoch: {err_list}")
```
