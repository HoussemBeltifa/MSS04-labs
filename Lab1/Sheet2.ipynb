# Truth tables and related functions

## Truth tables

```python
from sympy.logic.boolalg import truth_table

print(list(truth_table(x&y, [x,y])))
print()
table = truth_table(x>>y, [x,y])
for t in table:
    print(t)
```

## Interference

```python
from sympy.logic.inference import satisfiable

satisfiable(x & ~x)
```

```python
satisfiable(x | ~x)
```

```python
models=satisfiable(x | ~x, all_models=True) #get a generator of all models
for model in models:
    print(model)
```

```python
models=satisfiable(x | y, all_models=True) 
for model in models:
    print(model)
```

```python
models=satisfiable(x & ~x, all_models=True) 
for model in models:
    print(model)
```

```python
term = x & ~y
models=satisfiable(term, all_models=True) 
for model in models:
    if model:
        print(model)
    else:
        print("Unsatisfiable !")
```
