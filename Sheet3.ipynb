EX 9 (Activationfunctions)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
x = np.linspace(-2, 2, 500)
alpha = 0.1

activation_functions = {
    "Sigmoid": 1 / (1 + np.exp(-x)),
    "Tanh": 2 / (1 + np.exp(-2 * x)) - 1,
    "ReLU": np.maximum(0, x),
    "Leaky ReLU": np.where(x > 0, x, alpha * x),
    "SoftPlus": np.logaddexp(0, x),
    "ELU": np.where(x > 0, x, alpha * (np.exp(x) - 1))
}

fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.ravel()

for ax, (name, y) in zip(axes, activation_functions.items()):
    ax.plot(x, y)
    ax.set_title(name)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))

for name, y in activation_functions.items():
    plt.plot(x, y, label=name)

plt.title("Activation Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
